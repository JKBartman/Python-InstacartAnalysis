# Define Price Range Column

```python
# define price_lable function
def price_label(row):
  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
#define price_range column
df['price_range'] = df.apply(price_label, axis=1)
```

```python
df['price_range'].value_counts(dropna=False)
```

```python
df['prices'].min()
```

```python
ords_prods_merge.loc[ords_prods_merge['prices'] <= 5, 'price_range_loc'] ="Low-range product"
```

# Define Busiest and Slowest Days

```python
# 2. Create a new column that lables the 2 "Busiest days" and 2 "Slowest days"
```

```python
# determine busiest and slowest days
ords_prods_merge['orders_day_of_week'].value_counts(dropna=False)
```

```python
# day 0 and day 1 are the busiest days, day 3 and day 4 are the slowest days
```

```python
# Set 'result' to calculate busiest and slowest days
result = []

for value in ords_prods_merge["orders_day_of_week"]:
  if value == 0:
    result.append("Busiest days")
  elif value == 1:
    result.append("Busiest days") 
  elif value == 3:
    result.append("Slowest days")
  elif value == 4:
    result.append("Slowest days")
  else:
    result.append("Regularly busy")
```

```python
result
```

```python
# merge new 'busiest_days' column into df
```

```python
ords_prods_merge['busiest_days']=result
```

```python
# 3. Check values for accuracy
ords_prods_merge['busiest_days'].value_counts(dropna=False)
```

# Define busiest periods of the day column

```python
# 4. Create a new column 'busiest_period_of_day' by "Most orders", "Average orders", & "Fewest orders"
```

```python
# determine busiest and slowest days
```

```python
ords_prods_merge.head()
```

```python
ords_prods_merge['order_hour_of_day'].value_counts(dropna=False)
```

```python
# hours 10 & 11 are the Most orders, hours 3 & 4 are the Fewest orders 
```

```python
# Set 'result' to calculate most and fewest days
result = []

for value in ords_prods_merge["order_hour_of_day"]:
  if value == 10:
    result.append("Most Orders")
  elif value == 11:
    result.append("Most Orders") 
  elif value == 3:
    result.append("Fewest orders")
  elif value == 4:
    result.append("Fewest orders")
  else:
    result.append("Average orders")
```

```python
result
```

```python
# create new column and merge 'busiest_period_of_day' into df
ords_prods_merge['busiest_period_of_day']=result
```

```python
# Check values for accuracy
ords_prods_merge['busiest_period_of_day'].value_counts(dropna=False)
```
