# Create a Spender flag

```python
# 6. Create a spending flag for each user based on the average price across all their orders using the following criteria:
#      mean price < 10, flag as "Low spender"
#      mean price >= 10, flag as "High spender" 
```

```python
# create a column fom mean_spend
ords_prods_merge['mean_spend']=ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
ords_prods_merge.head()
```

```python
# Create flags for high and low spenders
ords_prods_merge.loc[ords_prods_merge['mean_spend']<10,'spender_flag']='Low_spender'
ords_prods_merge.loc[ords_prods_merge['mean_spend']>= 10, 'spender_flag'] = 'High_spender'
```

```python
# Check spender flags
ords_prods_merge['spender_flag'].value_counts(dropna = False)
```

# Create a Frequency flag

```python
# 7. Create an order frequency flag that marks the regularity of a user’s ordering behavior
#    according to the median in the “days_since_prior_order” column. The criteria for the flag should be as follows:
#      median of “days_since_last_order” > 20, then flag as “Non-frequent customer"
#      median of “days_since_last_order” > 10 & <= 20, then flag as “Regular customer"
#      median of “days_since_last_order” < 10, then flag as “Frequent customer"
```

```python
# create a column for frequent customers
ords_prods_merge['customer_frequency'] = ords_prods_merge.groupby(['user_id'])['days_since_last_order'].transform(np.median)
```

```python
# Creating flag on order frequency
ords_prods_merge.loc[ords_prods_merge['customer_frequency'] > 20, 'frequency_flag'] = 'Non_frequent_customer'
ords_prods_merge.loc[(ords_prods_merge['customer_frequency']>10) & (ords_prods_merge['customer_frequency']<=20),'frequency_flag'] = 'Regular customer'
ords_prods_merge.loc[ords_prods_merge['customer_frequency'] <= 10, 'frequency_flag'] = 'Frequent Customer'
```

```python
# Check frequency flags
ords_prods_merge['frequency_flag'].value_counts(dropna = False)
```
